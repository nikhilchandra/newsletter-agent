```python
import requests
import json
```

```python
# https://www.ncbi.nlm.nih.gov/books/NBK25500/
```

```python
# basic searching
# https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term=science[journal]+AND+breast+cancer+AND+2008[pdat]
url='https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi'
params = {
    'db':'pubmed',
    'term':'science[journal] AND breast cancer AND 2008[pdat]',
    'retmode':'json',
}
# where do i get an api key?
r = requests.get(url=url, params=params)
print(r.status_code)
```

```python
r.json()
```

```python
uid_list = ','.join(r.json()['esearchresult']['idlist'])
```

```python
# downloading basic summaries
# esummary.fcgi?db=<database>&id=<uid_list>

url='https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi'
params = {
    'db':'pubmed',
    'id':uid_list,
    'retmode':'json',    
}
r2 = requests.get(url=url, params=params)
```

```python
r2.json()
```

```python
r2.json()['result'].keys()
```

```python
# let's extract info for first paper, p
p = r2.json()['result']['19008416']

for key in p.keys():
    print(f'{key}: {p[key]}')
```
